```
table_4 = '''"Program Number" "Duration (days)" "Programs that interfere"
1 2 3 5 8
2 2 3 7 10
3 4 1 2 12
4 3 6 7 14
5 2 1 9 12
6 3 4 7 11 14
7 5 2 4 6
8 2 1 10 13
9 3 5 15
10 3 2 8
11 2 6 12 15
12 4 3 5 11
13 3 8 14
14 4 4 6 13
15 3 9 11'''

rows = table_4.split('\n')
head = rows[0]
data = rows[1:]

prog_num = list()
duration = list()
clashes = list()

for row in data:
  a, b, c = row.split(' ', 2)
  prog_num.append(a)
  duration.append(b)
  clashes.append(c)
```

```
skills = pd.read_table(StringIO(table_1), sep=' ')
jobs_a = pd.read_table(StringIO(table_2a), sep=' ')
jobs_b = pd.read_table(StringIO(table_2b), sep=' ')
prog_a = pd.read_table(StringIO(table_3a), sep=' ')
prog_b = pd.read_table(StringIO(table_3b), sep=' ')
```

```
skills = skills.sort_values('Number').reset_index(drop=True)
jobs = pd.merge(jobs_a, jobs_b)
prog = pd.merge(prog_a, prog_b)
```

```
clash_grid = [[str(i+1) in row for i in range(len(prog_num))] for row in clashes]
clash_grid = np.array(clash_grid) * 1
prog_clash = pd.DataFrame(clash_grid, index=prog_num, columns=prog_num)
prog_clash.index.name = 'Program'
prog_clash.to_csv('prog_clash.csv')
```

```
prog_skill = prog.drop(columns=['Enrollment Cost ($)'])
prog_skill = prog_skill.set_index('Program')
prog_skill.columns.name = 'Skill'
prog_skill.to_csv('prog_skill.csv')
```

```
prog_info = prog[['Program', 'Enrollment Cost ($)']].copy()
prog_info['Duration (days)'] = pd.to_numeric(duration)
prog_info = prog_info.set_index('Program')
prog_info.to_csv('prog_info.csv')
```
